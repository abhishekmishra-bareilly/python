**try to add some lables**

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine values')
plt.title('sine wave')
plt.show()
```

```python
plt.plot(x,z)
plt.xlabel('angle')
plt.ylabel('sine values')
plt.title('sine wave')
plt.show()
```

**parabola plot**

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x), 'g')
plt.plot(x,np.cos(x), 'r')
```

**Bar plot**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languase = ['English','french','spenish','germen']
pople = [100,150,50,120]
ax.bar(languase,pople,)
plt.xlabel('LANGUAGES')
plt.ylabel('PEOPLE')
plt.show()
```

**pie chart**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languase = ['English','french','spenish','germen']
pople = [100,150,50,120]
ax.pie(pople,labels =languase, autopct='%1.1f%%')
plt.show()
```

**scatter plot**

```python
x = np.linspace(0,10,30)
y= np.sin(x)
z= np.cos(x)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(x,y,color ='g')
ax.scatter(x,z,color = 'r')
plt.show()
```

**3D scatter plot**

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
z= 20*np.random.random(100)
x= np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
```
