#using boston house price data

```python
from sklearn.datasets import load_boston
house_bostern = load_boston()
```

```python
print(house_bostern)
```

```python
house = pd.DataFrame(house_bostern.data,columns = house_bostern.feature_names)
house['price'] = house_bostern.target
```

```python
house.head()
```

**Distribution plot**
